## Loading the saved model.......

```python
#import pickle file
import pickle

model_name = "bike_rental_model.pkl"
```

```python
#load the model
load_model = pickle.load(open(model_name, 'rb'))
print("Model Loaded Successfully")
```

```python
#predicting
import pandas as pd

cols = ["season", "yr", "mnth", "holiday", "weekday",
        "workingday", "weathersit", "temp", "atemp", "hum", "windspeed"]

values = [[3, 1, 7, 0, 4, 1, 1, 0.67, 0.65, 0.48, 0.19]]

predict = pd.DataFrame(values, columns=cols)

load_model.predict(predict)
```
